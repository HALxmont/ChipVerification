```python
import control
import numpy as np
import control.matlab as cmat
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
```

# Pixels vs current

```python
data = pd.read_excel("tesis_base.xlsx")
data.columns = ['x', 'Pixels', 'Charges[q]', 'Amplitude[mA]']
```

```python
data
```

```python
fig = px.line(data, x ='Pixels', y = 'Amplitude[mA]', 
              labels={
                     "Pixels": "Pixels",
                     "Amplitude[mA]": "Amplitude [mA]"
                 })

fig.layout.title.text = 'Amplitude vs Activated Pixels'
fig.update_layout(
    #showlegend=False,
    plot_bgcolor="#f5f5f5",
    #margin=dict(t=10,l=10,b=10,r=10)
    title_x=0.5,
    #title_font_family="Old Standard TT",
    #legend_font_family="Old Standard TT",
    font_family="Cambria",
    font_size = 25,
    title_font_size=40

)


fig.write_image("amplitude-vs-pixels.pdf",width=0.8*1500, height=0.3*1500, scale=1)
fig.show()
```

```python
fig = px.line(data, x ='Pixels', y = 'Charges[q]', 
              labels={
                     "Pixels": "Pixels",
                     "Charges[q]": "Charge [q]"
                 },)

fig.layout.title.text = 'Charge[q] vs Activated Pixels'
fig.update_layout(
    #showlegend=False,
    plot_bgcolor="#f5f5f5",
    #margin=dict(t=10,l=10,b=10,r=10)
    title_x=0.5,
    #title_font_family="Old Standard TT",
    #legend_font_family="Old Standard TT",
    font_family="Cambria",
    font_size = 25,
    title_font_size=40

    
)

fig.write_image("charge-vs-pixels.pdf",width=0.8*1500, height=0.3*1500, scale=1)
fig.show()
```
